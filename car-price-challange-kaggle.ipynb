###  Removing the old data (before 2010)
* We see that most of the cars are after 2010 and car industry has also changed in these years. 
* Training the model on new data will give more accurate results.

```python
cardfn = cardf[cardf.Year>2010]
```

```python
cardfn
```

```python
cardfn[['Present_Price', 'Kms_Driven','Selling_Price']].corr()
```

### Corrleation improved 
* after removing old data, corr bw Selling_Price and Present_Price improved to 90%

```python
for col in ['Fuel_Type','Seller_Type','Transmission',]:
    dummies = pd.get_dummies(cardfn[col], prefix=col)
    cardfn = pd.concat([cardfn, dummies], axis=1)    

cardfn['age'] = 2020 - cardfn.Year

dropcols = ['Car_Name', 'Year']+['Fuel_Type','Seller_Type','Transmission',]
cardfn.drop(dropcols, axis=1, inplace=True)
```

```python
cardfn
```

```python
sns.pairplot(cardfn)
```

### Train and test dataset

```python
X, y = cardfn.drop('Selling_Price', axis=1), cardfn.Selling_Price
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

### Scaling the input features 
* train the scaler on training dataset
* transform test data using trained scaler

```python
from joblib import dump, load
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import metrics
```

```python
model = Pipeline(verbose=1, steps=[
    ('scaler', StandardScaler()),
    ('polyfeats', PolynomialFeatures()),
    ('ridge', Ridge())
])
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('explained_variance_score = ', metrics.explained_variance_score(y_test, y_pred))
print('mean_squared_error = ', metrics.mean_squared_error(y_test, y_pred))
print(f'\033[1mr2_score = { metrics.r2_score(y_test, y_pred)}\033[0m')
```

### make test pipeline 

```python
val_dict = {
    'Year':2018,
    'Present_Price':5.7,
    'Kms_Driven':6500,
    'Fuel_Type':'Diesel',
    'Seller_Type':'Dealer',
    'Transmission':'Automatic',
    'Owner':0,
}

testdf = pd.DataFrame([val_dict])
for col in ['Fuel_Type','Seller_Type','Transmission',]:
    dummies = pd.get_dummies(testdf[col], prefix=col)
    testdf = pd.concat([testdf, dummies], axis=1)    
testdf['age'] = 2020 - testdf.Year

dropcols = ['Year']+['Fuel_Type','Seller_Type','Transmission',]
testdf.drop(dropcols, axis=1, inplace=True)

columns = ['Present_Price', 'Kms_Driven', 'Owner', 'Fuel_Type_CNG',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Dealer',
       'Seller_Type_Individual', 'Transmission_Automatic',
       'Transmission_Manual', 'age']
testdf = pd.concat([testdf, pd.DataFrame([], columns=columns)]).fillna(0)[columns]
```
